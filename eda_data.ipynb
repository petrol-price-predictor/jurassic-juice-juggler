```python
import pandas as pd
import json

def get_days(day_code):
    days = []
    days_dict = {
        1: "Monday",
        2: "Tuesday",
        4: "Wednesday",
        8: "Thursday",
        16: "Freiday",
        32: "Saturday",
        64: "Sunday",
        128: "Holiday"
    }
    for code, day in days_dict.items():
        if day_code & code:
            days.append(day)
    return days

def extract_opening_times(row):
    opening_times_json = json.loads(row['openingtimes_json'])
    opening_times = []
    if 'openingTimes' in opening_times_json:
        for period in opening_times_json['openingTimes']:
            days = get_days(period['applicable_days'])
            start_time = period['periods'][0]['startp']
            end_time = period['periods'][0]['endp']
            opening_times.append({
                'days': days,
                'start_time': start_time,
                'end_time': end_time
            })
    return opening_times

# Apply the function to each row of the dataframe
stations_data['opening_times'] = stations_data.apply(extract_opening_times, axis=1)

```

- In the following function, the **extract_opening_times** function creates an **opening_times** dictionary for each row of the input data frame with keys for each day of the week and holidays, and values for the start and end times of the working day specified by 'period'. The **opening_times** dictionary is returned by the function.

- The modified data frame is created by applying the **extract_opening_times** function to each row of the input data frame and converting the resulting **opening_times** dictionary into a separate data frame using the pd.Series constructor. The resulting data frames are concatenated horizontally using the pd.concat function and the original **openingtimes_json** column is dropped from the data frame using the drop method.

```python
import pandas as pd
import json

def get_days(day_code):
    days_dict = {
        1: "Monday",
        2: "Tuesday",
        4: "Wednesday",
        8: "Thursday",
        16: "Friday",
        32: "Saturday",
        64: "Sunday",
        128: "Holiday"
    }
    days = []
    for code, day in days_dict.items():
        if day_code & code:
            days.append(day)
    return days

def extract_opening_times(row):
    opening_times_json = json.loads(row['openingtimes_json'])
    opening_times = {
        'Monday': {'start_time': '', 'end_time': ''},
        'Tuesday': {'start_time': '', 'end_time': ''},
        'Wednesday': {'start_time': '', 'end_time': ''},
        'Thursday': {'start_time': '', 'end_time': ''},
        'Friday': {'start_time': '', 'end_time': ''},
        'Saturday': {'start_time': '', 'end_time': ''},
        'Sunday': {'start_time': '', 'end_time': ''},
        'Holiday': {'start_time': '', 'end_time': ''}
    }
    if 'openingTimes' in opening_times_json:
        for period in opening_times_json['openingTimes']:
            days = get_days(period['applicable_days'])
            for day in days:
                opening_times[day]['start_time'] = period['periods'][0]['startp']
                opening_times[day]['end_time'] = period['periods'][0]['endp']
    return opening_times

# Apply the function to each row of the dataframe
opening_times_df = stations_data.apply(lambda row: pd.Series(extract_opening_times(row)), axis=1)
stations_data = pd.concat([stations_data, opening_times_df], axis=1)
stations_data.drop(['openingtimes_json'], axis=1, inplace=True)
stations_data.drop(['opening_times'], axis=1, inplace=True)
```

- The following function would calculate the pay for the specified UUID **('e3972506-6546-4ba8-a597-e99187ae9c97')** on the specified day **('Monday')** based on the start and end times in the corresponding dictionary in the DataFrame. Note that this function assumes that the start and end times are in the format **'%H:%M'**, and it assumes a default hourly rate of 10 euros per hour, with higher rates for certain brands (ARAL and Shell). You may need to adjust the function to fit your specific use case.

```python
def calculate_payday(df, uuid, day):
    # Get the row for the specified UUID
    row = df[df['uuid'] == uuid].iloc[0]
    
    # Get the start and end times for the specified day
    hours = row[day]
    start_time = hours['start_time']
    end_time = hours['end_time']
    
    # Convert the times to datetime objects
    start_time = pd.to_datetime(start_time, format='%H:%M', errors='coerce')
    end_time = pd.to_datetime(end_time, format='%H:%M', errors='coerce')
    
    # Calculate the hours worked
    hours_worked = (end_time - start_time).total_seconds() / 3600
    
    # Calculate the hourly rate based on the brand (assuming a default rate of 10 euros per hour)
    hourly_rate = 10
    if row['brand'] == 'ARAL':
        hourly_rate = 12
    elif row['brand'] == 'Shell':
        hourly_rate = 15
    
    # Calculate the pay for the day
    pay = hours_worked * hourly_rate
    
    return pay
```

The following function returns a boolean value indicating whether the gas station is open at the specified time on the specified day.


```python
from datetime import datetime, timedelta

def is_station_open(uuid, day, hour):
    station = stations_data.loc[stations_data['uuid'] == uuid].iloc[0]
    start_time = station[day]['start_time']
    end_time = station[day]['end_time']
    if start_time == '' or end_time == '':
        return False
    start_hour, start_min = [int(x) for x in start_time.split(':')]
    end_hour, end_min = [int(x) for x in end_time.split(':')]
    opening_time = datetime.now().replace(hour=start_hour, minute=start_min, second=0, microsecond=0)
    closing_time = datetime.now().replace(hour=end_hour, minute=end_min, second=0, microsecond=0)
    if closing_time < opening_time:
        closing_time += timedelta(days=1)
    requested_time = datetime.now().replace(hour=hour, minute=0, second=0, microsecond=0)
    return opening_time <= requested_time <= closing_time


uuid = "e3972506-6546-4ba8-a597-e99187ae9c97"
day = "Monday"
hour = datetime.now().hour
print(is_station_open(uuid, day, hour))
```

This will return a new dataframe station_info_with_opening_times with the opening times added. Note that this function does not modify the original station_info dataframe, but instead returns a new one with the updates.

```python
stations_info = pd.read_csv('data/stations/stations.csv')
stations_data = pd.read_csv('data/stations/2023/05/2023-05-10-stations.csv')
```

```python
import pandas as pd
import json

def add_opening_times_to_station_info(station_info, station_data):
    def get_days(day_code):
        days_dict = {
            1: "Monday",
            2: "Tuesday",
            4: "Wednesday",
            8: "Thursday",
            16: "Friday",
            32: "Saturday",
            64: "Sunday",
            128: "Holiday"
        }
        days = []
        for code, day in days_dict.items():
            if day_code & code:
                days.append(day)
        return days

    def extract_opening_times(row):
        opening_times_json = json.loads(row['openingtimes_json'])
        opening_times = {
            'Monday': {'start_time': '', 'end_time': ''},
            'Tuesday': {'start_time': '', 'end_time': ''},
            'Wednesday': {'start_time': '', 'end_time': ''},
            'Thursday': {'start_time': '', 'end_time': ''},
            'Friday': {'start_time': '', 'end_time': ''},
            'Saturday': {'start_time': '', 'end_time': ''},
            'Sunday': {'start_time': '', 'end_time': ''},
            'Holiday': {'start_time': '', 'end_time': ''}
        }
        if 'openingTimes' in opening_times_json:
            for period in opening_times_json['openingTimes']:
                days = get_days(period['applicable_days'])
                for day in days:
                    opening_times[day]['start_time'] = period['periods'][0]['startp']
                    opening_times[day]['end_time'] = period['periods'][0]['endp']
        return opening_times
    
    # Create a new dataframe with the opening times for each station
    opening_times_df = station_data.apply(lambda row: pd.Series(extract_opening_times(row)), axis=1)
    opening_times_df.index = station_data['uuid']
    
    # Add the opening times to the station_info dataframe
    station_info = station_info.join(opening_times_df, on='uuid')
    
    # Remove the 'openingtimes_json' column from the station_info dataframe
    if 'openingtimes_json' in station_info.columns:
        station_info.drop('openingtimes_json', axis=1, inplace=True)

    
    return station_info
```

```python
station_info_with_opening_times = add_opening_times_to_station_info(stations_info, stations_data)
```
