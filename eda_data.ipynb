```python
#!pip install folium
import folium

# Create a map centered on the city
city = "Breitscheid-Rabenscheid"
location = (50.681303,	8.148122) # Latitude and longitude of Munich
map = folium.Map(location=location, zoom_start=12)

# Add a marker for each gas station in the city
for i, row in stations_data[stations_data["city"] == city].iterrows():
    folium.Marker([row["latitude"], row["longitude"]], popup=row["name"]).add_to(map)

# Display the map
map
```
